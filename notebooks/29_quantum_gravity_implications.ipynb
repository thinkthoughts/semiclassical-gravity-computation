## 6. Synthesis figure

```python
labels = [
    "Semiclassical gravity",
    "Schrödinger–Newton dynamics",
    "Nonlinear evolution",
    "Efficient NP-complete solutions",
    "PECTT tension",
    "Constraints on theory choice",
    "Quantum gravity implications"
]

fig, ax = plt.subplots(figsize=(7, 10))

x = [0] * len(labels)
y = list(range(len(labels), 0, -1))

ax.scatter(x, y, s=320, zorder=3)

for yi, label in zip(y, labels):
    ax.text(
        0,
        yi + 0.3,
        label,
        ha="center",
        va="center",
        fontsize=11,
        fontweight="bold"
    )

for yi in y[:-1]:
    ax.annotate(
        "",
        xy=(0, yi - 0.72),
        xytext=(0, yi - 0.28),
        arrowprops={"arrowstyle": "->", "lw": 2}
    )

ax.set_xlim(-1.4, 1.4)
ax.set_ylim(0.5, len(labels) + 0.7)
ax.axis("off")

ax.set_title(
    "Computational implications for quantum gravity",
    fontsize=15,
    pad=20
)

figure_path = FIGURES / "29_quantum_gravity_implications.png"

fig.savefig(
    figure_path,
    dpi=200,
    bbox_inches="tight"
)

plt.show()

figure_path
```

## 11. Export artifacts

```python
chain_csv = RESULTS / "29_argument_chain.csv"
interpretations_csv = RESULTS / "29_interpretations.csv"
future_csv = RESULTS / "29_future_directions.csv"
md_path = RESULTS / "29_quantum_gravity_implications.md"

argument_chain.to_csv(chain_csv, index=False)
interpretations.to_csv(interpretations_csv, index=False)
future_directions.to_csv(future_csv, index=False)

report = []

report.append("# 29 Quantum Gravity Implications")
report.append("")
report.append("## Notebook chain")

for _, row in argument_chain.iterrows():
    report.append(
        f"- **{row['notebook']}** — {row['contribution']}"
    )

report.append("")
report.append("## Candidate interpretations")

for _, row in interpretations.iterrows():
    report.append(
        f"- **{row['interpretation']}** — {row['motivation']}"
    )

report.append("")
report.append("## Engineering statement")

for key, value in engineering_statement.items():
    report.append(
        f"- **{key.replace('_',' ').title()}**: {value}"
    )

report.append("")
report.append("## Bottom line")
report.append(bottom_line)

report.append("")
report.append("## Future directions")

for _, row in future_directions.iterrows():
    report.append(
        f"- **{row['direction']}** — {row['question']}"
    )

md_path.write_text(
    "\n".join(report),
    encoding="utf-8"
)

print(f"Saved {chain_csv}")
print(f"Saved {interpretations_csv}")
print(f"Saved {future_csv}")
print(f"Saved {md_path}")
print(f"Saved {figure_path}")
```
