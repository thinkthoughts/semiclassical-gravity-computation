## 6. Revised synthesis figure

The first version ended with a single phrase: "quantum gravity implications."

This version branches after "constraints on theory choice" to show that the computational argument narrows the interpretive space without uniquely determining one answer.

```python
pipeline = [
    "Semiclassical gravity",
    "Schrödinger–Newton dynamics",
    "Nonlinear evolution",
    "Efficient NP-complete solutions",
    "PECTT tension",
    "Constraints on theory choice"
]

branches = [
    "Semiclassical gravity\nincomplete",
    "Gravity\nfundamentally quantum",
    "Alternative\ncouplings",
    "Hidden\nlimitations"
]

fig, ax = plt.subplots(figsize=(9, 11))
ax.axis("off")

ys = list(range(len(pipeline) + 2, 2, -1))

for y, label in zip(ys, pipeline):
    ax.scatter([0], [y], s=280, zorder=3)
    ax.text(
        0,
        y + 0.32,
        label,
        ha="center",
        va="center",
        fontsize=11,
        fontweight="bold"
    )

for y1, y2 in zip(ys[:-1], ys[1:]):
    ax.annotate(
        "",
        xy=(0, y2 + 0.42),
        xytext=(0, y1 - 0.42),
        arrowprops={"arrowstyle": "->", "lw": 2}
    )

branch_y = 1
branch_x = [-3.0, -1.0, 1.0, 3.0]

for x, label in zip(branch_x, branches):
    ax.scatter([x], [branch_y], s=260, zorder=3)
    ax.text(
        x,
        branch_y - 0.55,
        label,
        ha="center",
        va="top",
        fontsize=10,
        fontweight="bold"
    )
    ax.annotate(
        "",
        xy=(x, branch_y + 0.32),
        xytext=(0, ys[-1] - 0.42),
        arrowprops={"arrowstyle": "->", "lw": 1.6}
    )

ax.set_xlim(-3.8, 3.8)
ax.set_ylim(0.0, len(pipeline) + 2.8)

ax.set_title(
    "Computational constraints and quantum gravity interpretations",
    fontsize=15,
    pad=20
)

figure_path = FIGURES / "29_quantum_gravity_implications.png"

fig.savefig(
    figure_path,
    dpi=200,
    bbox_inches="tight"
)

plt.show()

figure_path
```
