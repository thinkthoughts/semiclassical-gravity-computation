# 13 Schrödinger–Newton Dynamics

This notebook follows the paper's route from semiclassical gravity toward Schrödinger–Newton-type dynamics.

The goal is not to solve the equation numerically. The goal is to identify where the nonlinear state-dependence enters:

`semiclassical Einstein equation → weak-field approximation → Newtonian potential → quantum mass density → Schrödinger–Newton dynamics`

## 3. Semiclassical pathway

The key idea is that classical gravity is sourced by quantum matter.

In semiclassical gravity, the gravitational side is classical while the matter source is represented by an expectation value.

```python
semiclassical_pathway = pd.DataFrame([
    {
        "step": 1,
        "stage": "Semiclassical Einstein equation",
        "working_role": "Classical spacetime curvature responds to quantum matter expectation values."
    },
    {
        "step": 2,
        "stage": "Weak-field approximation",
        "working_role": "The relativistic gravitational field is approximated by a Newtonian potential."
    },
    {
        "step": 3,
        "stage": "Newtonian gravitational potential",
        "working_role": "The potential is determined by a mass density source."
    },
    {
        "step": 4,
        "stage": "Quantum mass density",
        "working_role": "The source depends on the quantum state."
    },
    {
        "step": 5,
        "stage": "Schrödinger–Newton dynamics",
        "working_role": "The quantum state evolves under a potential generated by that same state."
    },
    {
        "step": 6,
        "stage": "Nonlinear state-dependent evolution",
        "working_role": "The evolution depends on the state being evolved."
    }
])

semiclassical_pathway
```

## 4. Equation-level sketch

This notebook keeps the math at the level needed for the report.

The route is:

1. Semiclassical Einstein equation
2. Weak-field limit
3. Poisson equation for a Newtonian potential
4. Schrödinger equation with a potential sourced by the quantum state

The important feature is not a numerical solution. The important feature is the dependence:

\[
V_\psi = V[\psi]
\]

so the evolution of \(\psi\) depends on \(\psi\) itself.

```python
equation_sketch = pd.DataFrame([
    {
        "name": "Semiclassical Einstein equation",
        "schematic_form": "G = expectation value of T",
        "role": "Couples classical gravity to quantum matter."
    },
    {
        "name": "Poisson equation",
        "schematic_form": "∇²Φ = source density",
        "role": "Weak-field Newtonian gravitational potential."
    },
    {
        "name": "Quantum source",
        "schematic_form": "source density depends on |ψ|²",
        "role": "The quantum state sources the potential."
    },
    {
        "name": "Schrödinger–Newton equation",
        "schematic_form": "i∂ψ/∂t = H[ψ]ψ",
        "role": "The Hamiltonian depends on the state."
    }
])

equation_sketch
```

## 5. Where the nonlinearity enters

Standard linear quantum mechanics evolves by a Hamiltonian independent of the state being evolved.

Schrödinger–Newton-type dynamics are different because the gravitational potential is sourced by the quantum state itself.

That is the state-dependence relevant to the computational argument.

```python
nonlinearity_map = pd.DataFrame([
    {
        "case": "Linear quantum mechanics",
        "evolution": "i∂ψ/∂t = Hψ",
        "state_dependence": "H is fixed externally",
        "computational_relevance": "Standard BQP-style baseline."
    },
    {
        "case": "Schrödinger–Newton-type dynamics",
        "evolution": "i∂ψ/∂t = H[ψ]ψ",
        "state_dependence": "H depends on ψ through the gravitational potential",
        "computational_relevance": "Nonlinearity can change computational power."
    }
])

nonlinearity_map
```

## 6. Pathway figure

This figure records the conceptual route for later notebooks.

```python
labels = semiclassical_pathway["stage"].tolist()
x = list(range(1, len(labels) + 1))
y = [1] * len(labels)

fig, ax = plt.subplots(figsize=(13, 3.5))

ax.plot(x, y, marker="o", linewidth=2)

for xi, label in zip(x, labels):
    ax.text(
        xi,
        1.08,
        label,
        ha="center",
        va="bottom",
        fontsize=9,
        wrap=True
    )

for xi in x[:-1]:
    ax.annotate(
        "",
        xy=(xi + 0.78, 1),
        xytext=(xi + 0.22, 1),
        arrowprops={"arrowstyle": "->", "lw": 1.5}
    )

ax.set_xlim(0.6, len(labels) + 0.4)
ax.set_ylim(0.85, 1.25)
ax.axis("off")
ax.set_title("Weak-field route to Schrödinger–Newton nonlinearity", pad=20)

figure_path = FIGURES / "13_schrodinger_newton_pathway.png"
fig.savefig(figure_path, dpi=200, bbox_inches="tight")
plt.show()

figure_path
```
