## 4. Computational architecture

This figure records the conceptual pathway reconstructed from the computational argument.

```python
labels = ingredients["ingredient"].tolist()

fig, ax = plt.subplots(figsize=(6, 8))

x = [0] * len(labels)
y = list(range(len(labels), 0, -1))

ax.scatter(x, y, s=320, zorder=3)

for yi, label in zip(y, labels):
    ax.text(
        0,
        yi + 0.35,
        label,
        ha="center",
        va="center",
        fontsize=12,
        fontweight="bold"
    )

for yi in y[:-1]:
    ax.annotate(
        "",
        xy=(0, yi - 0.72),
        xytext=(0, yi - 0.28),
        arrowprops={"arrowstyle": "->", "lw": 2}
    )

ax.set_xlim(-1, 1)
ax.set_ylim(0.5, len(labels) + 0.75)
ax.axis("off")

ax.set_title(
    "Computational architecture of the Bao–Bouland–Jordan argument",
    fontsize=15,
    pad=20
)

ax.text(
    0,
    0.75,
    "Conceptual pathway reconstructed from the computational argument.",
    ha="center",
    fontsize=10
)

figure_path = FIGURES / "23_bao_bouland_jordan.png"

fig.savefig(
    figure_path,
    dpi=200,
    bbox_inches="tight"
)

plt.show()

figure_path
```

## 5. From computation to physics

Notebook 23 bridges computational structure and physical interpretation.

The argument becomes relevant to semiclassical gravity only when computational consequences are treated as evidence constraining physical theories.

```python
bridge = pd.DataFrame([
    {
        "computational": "State dependence",
        "physical": "Schrödinger–Newton dynamics"
    },
    {
        "computational": "Distinguishability amplification",
        "physical": "Computational consequences"
    },
    {
        "computational": "Decision extraction",
        "physical": "PECTT tension"
    },
    {
        "computational": "Efficient NP-complete solutions",
        "physical": "Constraints on theory choice"
    }
])

bridge
```

```python
fig, ax = plt.subplots(figsize=(9, 4))

ax.axis("off")

for i, (_, row) in enumerate(bridge.iterrows()):
    y = len(bridge) - i

    ax.text(
        0.2,
        y,
        row["computational"],
        fontsize=11,
        fontweight="bold",
        ha="center"
    )

    ax.text(
        0.8,
        y,
        row["physical"],
        fontsize=11,
        ha="center"
    )

    ax.annotate(
        "",
        xy=(0.65, y),
        xytext=(0.35, y),
        arrowprops={"arrowstyle": "->", "lw": 1.5}
    )

ax.text(0.2, len(bridge) + 0.6, "COMPUTATIONAL", ha="center", fontsize=12, fontweight="bold")
ax.text(0.8, len(bridge) + 0.6, "PHYSICAL", ha="center", fontsize=12, fontweight="bold")

bridge_figure_path = FIGURES / "23_computational_to_physical.png"

fig.savefig(
    bridge_figure_path,
    dpi=200,
    bbox_inches="tight"
)

plt.show()

bridge_figure_path
```

## 10. Export artifacts

```python
ingredients_csv = RESULTS / "23_computational_ingredients.csv"
assumptions_csv = RESULTS / "23_assumptions.csv"
bridge_csv = RESULTS / "23_computational_to_physical.csv"
md_path = RESULTS / "23_bao_bouland_jordan.md"

ingredients.to_csv(ingredients_csv, index=False)
assumptions.to_csv(assumptions_csv, index=False)
bridge.to_csv(bridge_csv, index=False)

report = []

report.append("# 23 Bao–Bouland–Jordan")
report.append("")
report.append("## Computational ingredients")

for _, row in ingredients.iterrows():
    report.append(f"- **{row['ingredient']}** — {row['role']}")

report.append("")
report.append("## Computational → Physical bridge")

for _, row in bridge.iterrows():
    report.append(
        f"- **{row['computational']}** → {row['physical']}"
    )

report.append("")
report.append("## Assumptions")

for _, row in assumptions.iterrows():
    report.append(
        f"- **{row['assumption']}** — {row['why_it_matters']}"
    )

report.append("")
report.append("## Bottom line")
report.append(bottom_line)

md_path.write_text("\n".join(report), encoding="utf-8")

print(f"Saved {ingredients_csv}")
print(f"Saved {assumptions_csv}")
print(f"Saved {bridge_csv}")
print(f"Saved {md_path}")
print(f"Saved {figure_path}")
print(f"Saved {bridge_figure_path}")
```
