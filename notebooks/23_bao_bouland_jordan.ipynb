## 4. Computational ingredients

## 5. Argument pathway

The figure below records the computational architecture of the argument.

```python
pathway = ingredients.copy()

fig, ax = plt.subplots(figsize=(8, 8))

x = [0] * len(pathway)
y = list(range(len(pathway), 0, -1))

ax.scatter(x, y, s=300, zorder=3)

for yi, (_, row) in zip(y, pathway.iterrows()):
    ax.text(
        -0.15,
        yi,
        row["ingredient"],
        ha="right",
        va="center",
        fontsize=12,
        fontweight="bold"
    )

    wrapped = "\n".join(wrap(row["role"], width=52))

    ax.text(
        0.15,
        yi,
        wrapped,
        ha="left",
        va="center",
        fontsize=10
    )

for yi in y[:-1]:
    ax.annotate(
        "",
        xy=(0, yi - 0.72),
        xytext=(0, yi - 0.28),
        arrowprops={"arrowstyle": "->", "lw": 1.8}
    )

ax.set_xlim(-2.5, 3.0)
ax.set_ylim(0.4, len(pathway) + 0.6)
ax.axis("off")
ax.set_title(
    "Computational architecture of the Bao–Bouland–Jordan argument",
    pad=18,
    fontsize=15
)

figure_path = FIGURES / "23_bao_bouland_jordan.png"

fig.savefig(
    figure_path,
    dpi=200,
    bbox_inches="tight"
)

plt.show()

figure_path
```

## 11. Export artifacts

```python
ingredients_csv = RESULTS / "23_computational_ingredients.csv"
assumptions_csv = RESULTS / "23_assumptions.csv"
md_path = RESULTS / "23_bao_bouland_jordan.md"

ingredients.to_csv(ingredients_csv, index=False)
assumptions.to_csv(assumptions_csv, index=False)

lines = []

lines.append("# 23 Bao–Bouland–Jordan\n")

lines.append("## Purpose\n")
lines.append(
    "This notebook identifies the computational ingredients "
    "connecting nonlinear quantum evolution to efficient "
    "NP-complete consequences.\n"
)

lines.append("## Computational ingredients\n")

for _, row in ingredients.iterrows():
    lines.append(
        f"- **{row['ingredient']}** — {row['role']}"
    )

lines.append("\n## Assumptions\n")

for _, row in assumptions.iterrows():
    lines.append(
        f"- **{row['assumption']}** — {row['why_it_matters']}"
    )

lines.append("\n## Engineering statement\n")

for key, value in engineering_statement.items():
    lines.append(
        f"- **{key.replace('_',' ').title()}**: {value}"
    )

lines.append("\n## Bottom line\n")
lines.append(bottom_line)

lines.append("\n## Reading questions\n")

for _, row in reading_questions.iterrows():
    lines.append(f"- {row['question']}")

md_path.write_text(
    "\n".join(lines),
    encoding="utf-8"
)

print(f"Saved {ingredients_csv}")
print(f"Saved {assumptions_csv}")
print(f"Saved {md_path}")
print(f"Saved {figure_path}")
```
