# 31 Toy Discrimination

This notebook adds a lightweight numerical demo after the conceptual sequence.

The goal is not to simulate semiclassical gravity or reproduce Bao–Bouland–Jordan.

The goal is to make one idea concrete:

> Nonlinear state-dependent evolution can amplify small differences between nearby states.

We compare three toy trajectories:

1. **Linear baseline** — distinguishability remains approximately unchanged.
2. **Toy nonlinear amplification** — distinguishability grows.
3. **Toy stochastic/noisy comparison** — distinguishability fluctuates without systematic computational amplification.

## 3. Toy model

We use a scalar distinguishability variable \(d(t)\) between two nearby states.

This is **not** a quantum gravity simulation.

It is a pedagogical toy model for the concept of distinguishability amplification.

- Linear baseline: \(d(t) = d_0\)
- Toy nonlinear amplification: \(d(t) = \frac{d_0 e^{kt}}{1 + d_0(e^{kt}-1)}\)
- Toy stochastic comparison: \(d(t)\) undergoes bounded random fluctuations without systematic amplification

```python
rng = np.random.default_rng(260614806)

t = np.linspace(0, 10, 201)
d0 = 0.02
k = 0.65

linear = np.full_like(t, d0)

nonlinear = (d0 * np.exp(k * t)) / (1 + d0 * (np.exp(k * t) - 1))

noise = np.empty_like(t)
noise[0] = d0
sigma = 0.006

for i in range(1, len(t)):
    step = rng.normal(0, sigma)
    pullback = -0.06 * (noise[i-1] - d0)
    noise[i] = noise[i-1] + step + pullback
    noise[i] = np.clip(noise[i], 0, 1)

data = pd.DataFrame({
    "time": t,
    "linear_baseline": linear,
    "toy_nonlinear_amplification": nonlinear,
    "toy_stochastic_comparison": noise
})

data.head()
```

## 4. Plot distinguishability growth

The key visual contrast:

- linear baseline stays near the initial distinguishability,
- nonlinear toy dynamics amplify distinguishability,
- stochastic comparison fluctuates without reliable amplification.

```python
fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(data["time"], data["linear_baseline"], label="Linear baseline")
ax.plot(data["time"], data["toy_nonlinear_amplification"], label="Toy nonlinear amplification")
ax.plot(data["time"], data["toy_stochastic_comparison"], label="Toy stochastic comparison", alpha=0.8)

ax.set_xlabel("Toy time")
ax.set_ylabel("Distinguishability")
ax.set_title("Toy distinguishability amplification")
ax.set_ylim(0, 1.05)
ax.legend()

figure_path = FIGURES / "31_distinguishability_growth.png"
fig.savefig(figure_path, dpi=200, bbox_inches="tight")
plt.show()

figure_path
```

## 5. Threshold crossing

A decision procedure becomes useful only if differences become operationally separable.

Here we define a toy threshold and ask when each trajectory crosses it.

```python
threshold = 0.5

def first_crossing(series, times, threshold):
    idx = np.where(series >= threshold)[0]
    if len(idx) == 0:
        return None
    return float(times[idx[0]])

crossings = pd.DataFrame([
    {
        "trajectory": "Linear baseline",
        "first_crossing_time": first_crossing(linear, t, threshold)
    },
    {
        "trajectory": "Toy nonlinear amplification",
        "first_crossing_time": first_crossing(nonlinear, t, threshold)
    },
    {
        "trajectory": "Toy stochastic comparison",
        "first_crossing_time": first_crossing(noise, t, threshold)
    }
])

crossings
```
