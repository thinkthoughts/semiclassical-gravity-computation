## 5. Visual map

This figure is intentionally simple. It avoids claiming unknown complexity class equalities.

The point is the paper's pathway:

`semiclassical gravity → nonlinearity → NP-complete power → PECTT tension`

```python
pathway = pd.DataFrame([
    {"step": 1, "label": "Semiclassical gravity"},
    {"step": 2, "label": "Nonlinear dynamics"},
    {"step": 3, "label": "Efficient NP-complete power"},
    {"step": 4, "label": "PECTT tension"}
])

fig, ax = plt.subplots(figsize=(10, 3))

x = pathway["step"].tolist()
y = [1] * len(x)
labels = pathway["label"].tolist()

ax.plot(x, y, marker="o", linewidth=2)

for xi, label in zip(x, labels):
    ax.text(xi, 1.06, label, ha="center", va="bottom", fontsize=10)

for xi in x[:-1]:
    ax.annotate(
        "",
        xy=(xi + 0.8, 1),
        xytext=(xi + 0.2, 1),
        arrowprops={"arrowstyle": "->", "lw": 1.5}
    )

ax.set_xlim(0.6, 4.4)
ax.set_ylim(0.85, 1.25)
ax.axis("off")
ax.set_title("Computational pathway in the lab report", pad=20)

figure_path = FIGURES / "complexity_pathway.png"
fig.savefig(figure_path, dpi=200, bbox_inches="tight")
plt.show()

figure_path
```

## 8. Export artifacts

This notebook exports:

- `results/07_complexity_classes.csv`
- `results/07_framework_comparison.csv`
- `results/07_complexity_classes.md`
- `figures/complexity_pathway.png`

```python
classes_csv = RESULTS / "07_complexity_classes.csv"
comparison_csv = RESULTS / "07_framework_comparison.csv"
md_path = RESULTS / "07_complexity_classes.md"

complexity_classes.to_csv(classes_csv, index=False)
comparison.to_csv(comparison_csv, index=False)

lines = []
lines.append("# 07 Complexity Classes\n")

lines.append("## Purpose\n")
lines.append(
    "This notebook establishes the working computational vocabulary for interpreting "
    "arXiv:2606.14806. The key issue is whether semiclassical gravity grants efficient "
    "NP-complete problem-solving power through nonlinear dynamics.\n"
)

lines.append("## Complexity classes\n")
for _, row in complexity_classes.iterrows():
    lines.append(f"- **{row['class']}** — {row['meaning']} _Role:_ {row['role_in_report']}")

lines.append("\n## Framework comparison\n")
for _, row in comparison.iterrows():
    lines.append(
        f"- **{row['framework']}** — typical class: {row['typical_class']}; "
        f"efficient NP-complete power: {row['efficient_np_complete_power']}; "
        f"role: {row['report_role']}."
    )

lines.append("\n## PECTT\n")
lines.append(f"- **Statement:** {pectt['working_statement']}")
lines.append(f"- **Role:** {pectt['role']}")

lines.append("\n## Paper-specific pathway\n")
for _, row in pathway.iterrows():
    lines.append(f"{row['step']}. {row['label']}")

lines.append("\n## Reading questions\n")
for _, row in reading_questions.iterrows():
    lines.append(f"- **{row['notebook']}** — {row['question']}")

md_path.write_text("\n".join(lines), encoding="utf-8")

print(f"Saved {classes_csv}")
print(f"Saved {comparison_csv}")
print(f"Saved {md_path}")
print(f"Saved {figure_path}")
```
