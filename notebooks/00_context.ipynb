# 00 Context — Computation Demystifies Semiclassical Gravity

This notebook anchors the repo for arXiv:2606.14806.

It loads `paper.yaml`, defines the working assumptions and vocabulary, maps the core argument, and exports a reusable context report to:

`results/00_context.md`

## 1. Setup paths

This cell works from either the repo root or the `notebooks/` directory.

```python
from pathlib import Path
import yaml
import pandas as pd

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

RESULTS = ROOT / "results"
FIGURES = ROOT / "figures"

RESULTS.mkdir(exist_ok=True)
FIGURES.mkdir(exist_ok=True)

print(f"ROOT    = {ROOT}")
print(f"RESULTS = {RESULTS}")
print(f"FIGURES = {FIGURES}")
```

## 2. Load `paper.yaml`

Expected location:

`paper.yaml`

If the file is missing, this notebook creates a minimal fallback dictionary so the context work can still run.

```python
paper_path = ROOT / "paper.yaml"

if paper_path.exists():
    with paper_path.open("r", encoding="utf-8") as f:
        paper = yaml.safe_load(f)
else:
    paper = {
        "title": "Semiclassical Gravity Efficiently Solves NP",
        "arxiv": "2606.14806",
        "category": "gr-qc",
        "html": "https://arxiv.org/html/2606.14806v1",
        "pdf": "https://arxiv.org/pdf/2606.14806",
        "note": "Fallback metadata because paper.yaml was not found."
    }

paper
```

## 3. Paper metadata

```python
metadata = {
    "title": paper.get("title"),
    "arxiv": paper.get("arxiv"),
    "category": paper.get("category"),
    "html": paper.get("html"),
    "pdf": paper.get("pdf"),
}

metadata_df = pd.DataFrame([metadata])
metadata_df
```

## 4. Core assumptions

The lab report begins by making the assumptions explicit.

```python
assumptions = pd.DataFrame([
    {
        "assumption": "Gravity is classical",
        "role": "Keeps the gravitational field outside quantum dynamics."
    },
    {
        "assumption": "Matter is quantum",
        "role": "Quantum states source the gravitational field."
    },
    {
        "assumption": "Coupling is semiclassical",
        "role": "Classical gravity responds to quantum expectation values."
    },
    {
        "assumption": "Weak-field limit applies",
        "role": "Connects semiclassical Einstein dynamics to Schrödinger–Newton dynamics."
    },
    {
        "assumption": "Physical computation obeys observed limits",
        "role": "PECTT rules out efficient physical solutions to NP-complete problems."
    }
])

assumptions
```

## 5. Key terms

These definitions are intentionally short. Later notebooks can refine them.

```python
terms = pd.DataFrame([
    [
        "P",
        "Problems efficiently solvable by deterministic classical computation."
    ],
    [
        "BQP",
        "Problems efficiently solvable by bounded-error quantum computation."
    ],
    [
        "NP",
        "Problems whose proposed solutions can be efficiently verified."
    ],
    [
        "NP-complete",
        "Hardest problems in NP under efficient reductions."
    ],
    [
        "PECTT",
        "Physical Extended Church–Turing Thesis: physically realizable processes do not efficiently solve NP-complete problems."
    ],
    [
        "Schrödinger–Newton equation",
        "Weak-field semiclassical dynamics where a quantum state sources a classical gravitational potential."
    ],
    [
        "Nonlinear quantum dynamics",
        "Dynamics where evolution depends on the quantum state itself."
    ]
], columns=["term", "working_definition"])

terms
```

## 6. Argument map

This is the first pass. The later notebooks refine each link.

```python
argument_map = pd.DataFrame([
    ["1", "Semiclassical gravity couples quantum matter to classical gravity."],
    ["2", "In the weak-field limit, this leads toward Schrödinger–Newton-type dynamics."],
    ["3", "These dynamics are nonlinear."],
    ["4", "Nonlinear quantum dynamics can enable efficient solutions to NP-complete problems."],
    ["5", "Observed physical computation appears constrained by PECTT."],
    ["6", "Therefore, computational complexity constrains which gravity theories remain plausible."]
], columns=["step", "claim"])

argument_map
```

## 7. Initial questions

```python
questions = [
    "Where exactly does nonlinearity enter the semiclassical model?",
    "How does the weak-field approximation connect to Schrödinger–Newton dynamics?",
    "Which computational result does the paper rely on most directly?",
    "What does PECTT rule out physically?",
    "What experiments could distinguish semiclassical gravity from quantized gravity?"
]

questions_df = pd.DataFrame({"question": questions})
questions_df
```

## 8. Engineering statement

The engineering statement is the bridge from reading to action.

```python
engineering_statement = {
    "statement": "Physical theories should preserve observed computational constraints.",
    "observation": "Semiclassical gravity introduces nonlinear quantum dynamics.",
    "context": "Generic nonlinear quantum theories can efficiently solve NP-complete problems.",
    "next_step": "Determine experimentally whether gravity exhibits fundamentally quantum behavior or requires an alternative coupling mechanism."
}

pd.DataFrame([engineering_statement])
```

## 9. Export context report

The output is a lightweight Markdown artifact that can be committed to `results/`.

```python
lines = []

lines.append("# 00 Context — Computation Demystifies Semiclassical Gravity\n")

lines.append("## Paper\n")
for key, value in metadata.items():
    lines.append(f"- **{key}:** {value}")

lines.append("\n## Core assumptions\n")
for _, row in assumptions.iterrows():
    lines.append(f"- **{row['assumption']}** — {row['role']}")

lines.append("\n## Key terms\n")
for _, row in terms.iterrows():
    lines.append(f"- **{row['term']}** — {row['working_definition']}")

lines.append("\n## Argument map\n")
for _, row in argument_map.iterrows():
    lines.append(f"{row['step']}. {row['claim']}")

lines.append("\n## Initial questions\n")
for q in questions:
    lines.append(f"- {q}")

lines.append("\n## Engineering statement\n")
for key, value in engineering_statement.items():
    label = key.replace("_", " ").title()
    lines.append(f"- **{label}:** {value}")

output_path = RESULTS / "00_context.md"
output_path.write_text("\n".join(lines), encoding="utf-8")

print(f"Saved context report to {output_path}")
output_path
```
