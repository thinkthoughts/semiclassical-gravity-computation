## 3. Linear quantum evolution

Ordinary quantum mechanics is linear in the state.

For this notebook, the key baseline is:

\[
i\frac{\partial \psi}{\partial t} = H\psi
\]

where \(H\) is fixed independently of the state being evolved.

```python
linear_quantum = pd.DataFrame([
    {
        "feature": "evolution",
        "linear_quantum": "i∂ψ/∂t = Hψ"
    },
    {
        "feature": "Hamiltonian",
        "linear_quantum": "H is fixed independently of ψ"
    },
    {
        "feature": "state distinguishability",
        "linear_quantum": "Inner-product structure constrains distinguishability"
    },
    {
        "feature": "computational baseline",
        "linear_quantum": "BQP"
    },
    {
        "feature": "physical role",
        "linear_quantum": "Standard quantum theory"
    }
])

linear_quantum
```

## 4. Nonlinear quantum evolution

Schrödinger–Newton-type dynamics introduce state-dependence.

For this notebook, the key schematic form is:

\[
i\frac{\partial \psi}{\partial t} = H[\psi]\psi
\]

where the Hamiltonian depends on the state being evolved.

```python
nonlinear_quantum = pd.DataFrame([
    {
        "feature": "evolution",
        "nonlinear_quantum": "i∂ψ/∂t = H[ψ]ψ"
    },
    {
        "feature": "Hamiltonian",
        "nonlinear_quantum": "H depends on ψ"
    },
    {
        "feature": "state distinguishability",
        "nonlinear_quantum": "State-dependent evolution can amplify distinctions"
    },
    {
        "feature": "computational baseline",
        "nonlinear_quantum": "Can exceed standard BQP in generic nonlinear models"
    },
    {
        "feature": "physical role",
        "nonlinear_quantum": "Theory feature under investigation"
    }
])

nonlinear_quantum
```

## 5. Linear versus nonlinear comparison

This table becomes a reusable artifact for the report.

```python
linear_vs_nonlinear = linear_quantum.merge(nonlinear_quantum, on="feature")
linear_vs_nonlinear
```

## 6. Computational consequences

The paper-specific concern is not that every nonlinear equation is physically realized.

The concern is that generic nonlinear quantum evolution can have computational consequences far beyond standard quantum computation.

```python
computational_consequences = pd.DataFrame([
    {
        "step": 1,
        "stage": "State dependence",
        "role": "The evolution rule depends on the quantum state being evolved."
    },
    {
        "step": 2,
        "stage": "Nonlinear evolution",
        "role": "The map from initial state to later state is not linear."
    },
    {
        "step": 3,
        "stage": "Enhanced distinguishability",
        "role": "Small differences between states can be amplified by nonlinear dynamics."
    },
    {
        "step": 4,
        "stage": "Efficient NP-complete power",
        "role": "Generic nonlinear quantum models can permit efficient solutions to NP-complete problems."
    },
    {
        "step": 5,
        "stage": "Physical constraint",
        "role": "Observed computational limits make this a test of physical plausibility."
    }
])

computational_consequences
```

## 7. Consequence figure

This figure uses a two-track structure:

- linear quantum evolution as the baseline
- nonlinear quantum evolution as the computational warning signal

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Track coordinates
tracks = {
    "Linear QM": {
        "x": [1, 2, 3],
        "y": [2.2, 2.2, 2.2],
        "labels": ["Linear\nevolution", "Unitary\nstructure", "BQP\nbaseline"]
    },
    "Nonlinear QM": {
        "x": [1, 2, 3, 4],
        "y": [1.0, 1.0, 1.0, 1.0],
        "labels": ["State\ndependence", "Nonlinear\nevolution", "Enhanced\ndistinguishability", "NP-complete\npower"]
    }
}

for track_name, data in tracks.items():
    ax.plot(data["x"], data["y"], marker="o", linewidth=2)
    for xi, yi, label in zip(data["x"], data["y"], data["labels"]):
        ax.text(xi, yi + 0.12, label, ha="center", va="bottom", fontsize=10)
    for xi, yi in zip(data["x"][:-1], data["y"][:-1]):
        ax.annotate(
            "",
            xy=(xi + 0.75, yi),
            xytext=(xi + 0.25, yi),
            arrowprops={"arrowstyle": "->", "lw": 1.5}
        )

ax.text(0.25, 2.2, "Linear QM", ha="left", va="center", fontsize=12, fontweight="bold")
ax.text(0.25, 1.0, "Nonlinear QM", ha="left", va="center", fontsize=12, fontweight="bold")

ax.set_xlim(0, 4.5)
ax.set_ylim(0.55, 2.75)
ax.axis("off")
ax.set_title("Why nonlinearity changes the computational question", pad=18, fontsize=15)

figure_path = FIGURES / "17_nonlinearity_consequences.png"
fig.savefig(figure_path, dpi=200, bbox_inches="tight")
plt.show()

figure_path
```
